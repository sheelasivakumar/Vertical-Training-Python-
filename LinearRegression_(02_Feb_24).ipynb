```python
# Importing dataset into dataframe

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Python Vertical Training/hiring.csv")
df
```

```python
df.shape
```

```python
# Checking for Null Values
df.isnull().sum()
```

```python
df.columns
```

```python
# Handling Null Values
df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean(),inplace = True)
df
```

```python
df['experience'].fillna('zero',inplace = True)
df
```

```python
!pip install word2number
```

```python
# Converting the string number to digits
from word2number import w2n
df['experience'] = df['experience'].apply(w2n.word_to_num)
df
```

```python
# Splitting data into Input(X) and Output (Y)
X = df[['experience','test_score(out of 10)','interview_score(out of 10)']]
Y = df['salary($)']
```

```python
X
```

```python
Y
```

```python
# Train and Test Split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
print(f'X_train : {x_train.shape}, Y_train : {y_train.shape}\nX_test : {x_test.shape}, Y_test : {y_test.shape}')
```

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.values,y_train.values)
```

```python
res = model.predict(x_test.values)
res
```

```python
# Predictin User input

n = int(input("Enter no of Candidates : "))
for i in range(n):
  exp = int(input("Experience : "))
  test_score = float(input("Test score: "))
  int_score = int(input("Interview score: "))
  ans = model.predict([[exp,test_score,int_score]])
  print(f"Salary (Candidate {i+1}): {ans}")
```
