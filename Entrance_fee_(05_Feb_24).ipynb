```python
# Loading Dataset into DataFrame

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Python Vertical Training/Top Indian Places to Visit.csv")
df.head()
```

```python
df = df[['Type','Google review rating','Entrance Fee in INR','Number of google review in lakhs']]
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df['Type'].unique()
```

```python
df.drop(df[df['Entrance Fee in INR'] == 0].index, inplace = True)
df.head()
```

```python
df.shape
```

```python
df = pd.get_dummies(df,dtype='int')
```

```python
df.head()
```

```python
df.shape
```

```python
# Splitting of Features
X = df.drop(columns = ['Entrance Fee in INR'])
Y = df['Entrance Fee in INR']
```

```python
# Train and Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.33)
print(f"x_train : {x_train.shape}\ny_train : {y_train.shape}\nx_test : {x_test.shape}\ny_test : {y_test.shape}")
```

```python
# Model Training

from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(x_train.values,y_train.values)
```

```python
lin_model.score(x_train.values,y_train.values)
```

```python
lin_model.score(x_test.values,y_test.values)
```

```python
from sklearn.linear_model import Lasso

las_model = Lasso(alpha = 50,max_iter = 100, tol = 0.1)
las_model.fit(x_train.values,y_train.values)
```

```python
las_model.score(x_train.values,y_train.values)
```

```python
las_model.score(x_test.values,y_test.values)
```

```python
from sklearn.linear_model import Ridge

rig_model = Ridge(alpha = 50,max_iter = 100, tol = 0.1)
rig_model.fit(x_train.values,y_train.values)
```

```python
rig_model.score(x_train.values,y_train.values)
```

```python
rig_model.score(x_test.values,y_test.values)
```
