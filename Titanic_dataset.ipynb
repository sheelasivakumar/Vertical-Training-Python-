```python
# Loading Dataset into Dataframe

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Python Vertical Training/titanic.csv")
df.head()
```

```python
df.shape
```

```python
# Checking for null Values
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
# Survival rate doesn't depend on PassengerId, Name, Ticket, Cabin, Fare, Embarked, Pclass, SibSp
df = df.drop(columns = ['PassengerId','Name','Ticket','Cabin','Fare','Embarked','Pclass','SibSp','Parch'])
df.head()
```

```python
# Handliing Missing Values
df['Age'].fillna(df['Age'].mean(),inplace = True)
df.isnull().sum()
```

```python
df.info()
```

```python
# Splitting Features
X = df[["Sex","Age"]]
Y = df["Survived"]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Splitting into Train and Test Dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.3)
print(f"x_train : {x_train.shape}\ny_train : {y_train.shape}\nx_test : {x_test.shape}\ny_test : {y_test.shape}")
```

```python
x1_train = pd.get_dummies(x_train,dtype = 'int')
x1_train.head()
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x1_train.values,y_train.values)
```

```python
log_model_acc = log_model.score(x1_train.values,y_train.values)
print(f"Accuracy of Logistic Model : {log_model_acc}")
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dec_model = DecisionTreeClassifier()
dec_model.fit(x1_train.values,y_train.values)
```

```python
dec_model_acc = dec_model.score(x1_train.values,y_train.values)
print(f"Accuracy of Decision Tree Model : {dec_model_acc}")
```

```python
# Changing the Training and Testing Data size
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.15)
print(f"x_train : {x_train.shape}\ny_train : {y_train.shape}\nx_test : {x_test.shape}\ny_test : {y_test.shape}")
```
