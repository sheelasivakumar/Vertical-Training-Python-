```python
# Loading Dataset into DataFrame

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Python Vertical Training/NY-House-Dataset.csv")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# Checking for null Values

df.isnull().sum()
```

```python
df = df.drop(columns = ['LATITUDE','LONGITUDE','BROKERTITLE','FORMATTED_ADDRESS','MAIN_ADDRESS','ADDRESS','LONG_NAME'])
df.head()
```

```python
df.nunique()
```

```python
df = df.drop(columns = ['STATE','STREET_NAME'])
df.head()
```

```python
# str to numerical
df = pd.get_dummies(df,dtype='int')
```

```python
# Splitting features

X = df.drop(columns  = ['PRICE'])
Y = df['PRICE']
```

```python
# Training and test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2)
print(f"x_train : {x_train.shape}\ny_train : {y_train.shape}\nx_test : {x_test.shape}\ny_test : {y_test.shape}")
```

```python
# Model Training

from sklearn.linear_model import LinearRegression

log_model = LinearRegression()
log_model.fit(x_train,y_train)
```

```python
log_model.score(x_train,y_train)
```

```python
log_model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso

las_model = Lasso(alpha = 50,max_iter = 100, tol = 0.1)
las_model.fit(x_train,y_train)
```

```python
las_model.score(x_train,y_train)
```

```python
las_model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge

rid_model = Ridge(alpha = 50,max_iter = 100, tol = 0.1)
rid_model.fit(x_train,y_train)
```

```python
rid_model.score(x_train,y_train)
```

```python
rid_model.score(x_test,y_test)
```
