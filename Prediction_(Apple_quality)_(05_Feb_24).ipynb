```python
# Loading Dataset into DataFrame

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Python Vertical Training/apple_quality.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Checking for null Values
df.isnull().sum()
```

```python
# Removing Null Values
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df['Quality'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['Quality'] = encoder.fit_transform(df['Quality'])

df['Quality'].unique()
```

```python
df.head()  # Good = 1  Bad = 0
```

```python
df.drop(columns = ['A_id'],inplace = True)
df.head()
```

```python
df['Acidity'] = df['Acidity'].astype(float)
df.info()
```

```python
import numpy as np

df = np.abs(df)
df.head()
```

```python
# Splitting Features X and Y

X = df.drop(columns = ['Quality'])
Y = df['Quality']
```

```python
# Train and Test Split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.20)
print(f"x_train : {x_train.shape}\ny_train : {y_train.shape}\nx_test : {x_test.shape}\ny_test : {y_test.shape}")
```

```python
# Model Training

from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()
log_model.fit(x_train.values,y_train.values)
```

```python
log_model.score(x_train.values,y_train.values)
```

```python
log_model.score(x_test.values,y_test.values)
```

```python
from sklearn.tree import DecisionTreeClassifier

dec_model = DecisionTreeClassifier()
dec_model.fit(x_train,y_train)
```

```python
dec_model.score(x_train,y_train)
```

```python
dec_model.score(x_test,y_test)
```

##### **Note:**
Decision Model gives an Accuracy of 1 for Training Dataset but the accuracy for test dataset is low (0.58)
